```python
class Bankaccount: 
    def __init__(self): 
        self.balance=0
        print("*******************************************")
        print("           Welcome to Piggy Bank           ")
   
    def deposit(self): 
        amount=float(input("Enter amount to be Deposited: ")) 
        self.balance += amount 
        print("\n Amount Deposited:",amount) 

    def withdraw(self): 
        amount = float(input("Enter amount to be Withdrawed: ")) 
        if self.balance>=amount: 
            self.balance-=amount 
            print("\n You Withdrawed:", amount) 
        else: 
            print("\n Insufficient balance ") 

    def display(self): 
        print("\n Net Available Balance=",self.balance) 

s = Bankaccount() 
op='C'
while(op !='E'):
    print("*******************************************")
    print("           Welcome to Piggy Bank           ")  
    print("              D - Deposit                  ")
    print("              W - Withdraw                 ")
    print("              C - Check Balance            ")
    print("              E - QUIT                     ")
    print("*******************************************")
    print("Which operation to perform (D/W/C) : ")
    op=input("Input the corresponding letter : ")
    if op == 'D':
        s.deposit() 
        s.display()        
    elif op == 'W':
        s.withdraw() 
        s.display()        
    elif op == 'C':
        s.display()
    else: 
        print("Thank you !!!")
        break
```
